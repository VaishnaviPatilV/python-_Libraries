```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the tips datasets

tips = sns.load_dataset("tips")
print(tips.head())
```

```python
tips
```

```python
tips.columns
```

```python
sns.scatterplot(x = "total_bill", y = "tip", data = tips)
plt.title("totle bill vs tips")
plt.grid()
plt.show()
```

```python
# Line Plot
sns.lineplot(x ="total_bill", y = "tip", data = tips)
plt.title("Total bill vs Tips")
plt.grid()
plt.show()
```

```python
#bar plot
sns.barplot(x = "day", y ="total_bill", data = tips)
#
plt.title("Bar plot day vs total_bill")
plt.show()
```

```python
sns.histplot(tips['total_bill'], kde = False) 
#kde stands for kernel density estimation. 
#KDE this plot are used to the data using a continuous probability density curve in one or more dimensions.
plt.title("histplot of total_bill")
plt.show()
```

```python
sns.boxplot(x = "day", y= "total_bill", data = tips)
# A box-and-whisker plot, is a data visualization tool used to display the distribution of a dataset.

plt.title("Total Bill by Day(Box Plot)")
plt.show()
```

```python
sns.pairplot(tips, hue = "sex")

plt.show()
#that shows pairwise relationships between multiple columns in a dataset,especially useful for exploring numerical data.
```

```python
# selected the numeric column
selected_columns = ['total_bill', 'tip', 'size']

corr = tips[selected_columns].corr() # checking for the correlation -- it mainly works with numerical data

# Plot heatmap
sns.heatmap(corr, annot = True, cmap='YlGnBu')
plt.title("Correlation Heatmap (Selected Columns)")
plt.show()
#heatmap is like colorful table,that represents the magnitude of values is mstrix or 2D.
```

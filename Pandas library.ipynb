# EDA -- Exploratory data analysis

Pandas stands for Python Data Analysis Library.
It provides high-level data structures and tools to make working with structured data (like tables, Excel files, SQL databases, CSVs) fast and easy.
It is built on top of NumPy.

NumPy stands for Numerical Python.
It is an open-source library used to work with arrays (especially multidimensional arrays).
It provides a fast way to perform mathematical operations on large datasets.
